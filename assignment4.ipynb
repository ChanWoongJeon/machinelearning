0. Prepared

```python
import csv
list_train_x = [[],[],[],[]]
list_train_y = []
list_test_x = [[],[],[],[]]
list_test_y = []
with open('data_train.csv', newline='') as myfile:
    reader  = csv.reader(myfile, delimiter=',')
    ct = 1 
    for i in reader:
        i[0]=float(i[0])
        i[1]=float(i[1])
        i[2]=float(i[2])
        i[3]=float(i[3])
        list_train_x[0].append(float(1))
        list_train_x[1].append(i[0])
        list_train_x[2].append(i[1])
        list_train_x[3].append(i[2])
        list_train_y.append(i[3])
        ct += 1

with open('data_test.csv', newline='') as myfile:
    reader  = csv.reader(myfile, delimiter=',')
    ct = 1 
    for i in reader:
        i[0]=float(i[0])
        i[1]=float(i[1])
        i[2]=float(i[2])
        i[3]=float(i[3])
        list_test_x[0].append(float(1))
        list_test_x[1].append(i[0])
        list_test_x[2].append(i[1])
        list_test_x[3].append(i[2])
        list_test_y.append(i[3])
        ct += 1
```

```python
import numpy as np
X_train = np.array(list_train_x)
X_test = np.array(list_test_x)
y_train =np.array(list_train_y)
y_test = np.array(list_test_y)
y_train = y_train.reshape(1,-1)
y_test = y_test.reshape(1,-1)
```

1. Plot the estimated parameters using the training dataset

```python
import matplotlib.pyplot as plt

obj = []
theta = [[],[],[],[]]
learning_rate =0.00001
m = len(y_train[0])
W = np.random.rand(1,4)
for i in range(3000000):
    cost = np.sum((np.dot(W,X_train)-y_train)**2)/(2*m)
    for k in range(4):
        theta[k].append(W[0][k])
    obj.append(cost)
    W = W - learning_rate*(((np.dot(W,X_train)-y_train)*X_train).sum(axis=1))/m
    
plt.figure(figsize=(14, 7))
plt.axvline(x=0, color = 'r',linestyle=':')
plt.axhline(y=0, color = 'r',linestyle=':')
plt.title("assignment4-1(W)")
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(range(3000000),theta[0],color='k',label='theta0')
plt.plot(range(3000000),theta[1],color='r',label='theta1')
plt.plot(range(3000000),theta[2],color='g',label='theta2')
plt.plot(range(3000000),theta[3],color='b',label='theta3')
plt.show()
```

그래프에서  theta0,1,2,3의 수렴을 확실히 알 수 있다

```python
plt.figure(figsize=(14, 7))
plt.axvline(x=0, color = 'r',linestyle=':')
plt.axhline(y=0, color = 'r',linestyle=':')
plt.title("assignment4-1(W)")
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0,30000)
plt.plot(range(3000000),theta[0],color='k',label='theta0')
plt.plot(range(3000000),theta[1],color='r',label='theta1')
plt.plot(range(3000000),theta[2],color='g',label='theta2')
plt.plot(range(3000000),theta[3],color='b',label='theta3')
plt.show()
```

범위를 300000까지 줄여서 보면 보다 정확히 theta1,2,3이 수렴함을 알 수 있다

3. Plot the testing error using the testing dataset at every iteration of gradient descent until convergence

```python
import matplotlib.pyplot as plt

obj = []
theta = [[],[],[],[]]
learning_rate =0.00001
m = len(y_test[0])
W = np.random.rand(1,4)
for i in range(100000):
    cost = np.sum((np.dot(W,X_test)-y_test)**2)/(2*m)
    for k in range(4):
        theta[k].append(W[0][k])
    obj.append(cost)
    W = W - learning_rate*(((np.dot(W,X_test)-y_test)*X_test).sum(axis=1))/m
    
plt.figure(figsize=(14, 7))
plt.axvline(x=0, color = 'r',linestyle=':')
plt.axhline(y=0, color = 'r',linestyle=':')
plt.title("assignment4-3(cost function)")
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0,1000)
plt.plot(range(100000),obj,color='b',label='cost function')
```

범위를 1000으로 줄여서 보면 testset의 costfuction의 수렴함을 알 수 있다.

```python
plt.figure(figsize=(14, 7))
plt.axvline(x=0, color = 'r',linestyle=':')
plt.axhline(y=0, color = 'r',linestyle=':')
plt.title("assignment4-1(W)")
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0,100000)
plt.plot(range(100000),theta[0],color='k',label='theta0')
plt.plot(range(100000),theta[1],color='r',label='theta1')
plt.plot(range(100000),theta[2],color='g',label='theta2')
plt.plot(range(100000),theta[3],color='b',label='theta3')
plt.show()
```
