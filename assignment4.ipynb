```python
import csv
list_train_x = [[],[],[],[]]
list_train_y = []
list_test_x = [[],[],[],[]]
list_test_y = []
with open('data_train.csv', newline='') as myfile:
    reader  = csv.reader(myfile, delimiter=',')
    ct = 1 
    for i in reader:
        i[0]=float(i[0])
        i[1]=float(i[1])
        i[2]=float(i[2])
        i[3]=float(i[3])
        list_train_x[0].append(float(1))
        list_train_x[1].append(i[0])
        list_train_x[2].append(i[1])
        list_train_x[3].append(i[2])
        list_train_y.append(i[3])
        ct += 1

with open('data_test.csv', newline='') as myfile:
    reader  = csv.reader(myfile, delimiter=',')
    ct = 1 
    for i in reader:
        i[0]=float(i[0])
        i[1]=float(i[1])
        i[2]=float(i[2])
        i[3]=float(i[3])
        list_test_x[0].append(float(1))
        list_test_x[1].append(i[0])
        list_test_x[2].append(i[1])
        list_test_x[3].append(i[2])
        list_test_y.append(i[3])
        ct += 1
```
