임의지정 random function y = 2*x+5/범위 (-5,5)/number of random point is 300

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return 2*x+5
```

```python
x = np.random.rand(1, 300)*10-5
sigma = np.random.rand(1,300)-0.5
y = func(x)
y_pred = func(x)+sigma
```

```python
plt.figure(figsize=(10, 10))
plt.axvline(x=0, color = 'r',linestyle=':')
plt.axhline(y=0, color = 'r',linestyle=':')
plt.title("assignment2")
plt.xlabel('X')
plt.ylabel('Y')
t1 = np.arange(-5.0, 5.0, 0.01)
plt.plot(t1,func(t1),color = 'b')
plt.scatter(x,y_pred,color='k',s=3)
plt.show()
```
