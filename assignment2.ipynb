임의지정 random function y = 2*x+5/범위 (-5,5)/number of random point is 300

2-1 Input data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x):
    return 2*x+5
```

```python
x = np.random.rand(1, 300)*10-5
sigma = np.random.randn(1,300)
y= func(x)+sigma
```

```python
plt.figure(figsize=(10, 10))
plt.axvline(x=0, color = 'r',linestyle=':')
plt.axhline(y=0, color = 'r',linestyle=':')
plt.title("assignment2-1")
plt.xlabel('X')
plt.ylabel('Y')
t1 = np.arange(-5.0, 5.0, 0.01)
plt.plot(t1,func(t1),color = 'b')
plt.scatter(x,y,color='k',s=3)
plt.show()
```

2-2 Output results

```python
theta = np.random.randn(1,2)
print(theta)
```

초기 랜덤으로 직선 구했을때(red-line)/blue-line은 원래의 직선

```python
plt.figure(figsize=(10, 10))
plt.axvline(x=0, color = 'r',linestyle=':')
plt.axhline(y=0, color = 'r',linestyle=':')
plt.title("assignment2-2")
plt.xlabel('X')
plt.ylabel('Y')
t1 = np.arange(-5.0, 5.0, 0.01)
plt.plot(t1,func(t1),color = 'b')
plt.scatter(x,y,color='k',s=3)
plt.plot(t1,theta[0,0]*t1+theta[0,1],color='r')
plt.show()
```

```python
obj = []
for i in range(1000):
    y_pred=theta[0,0]*x+theta[0,1]
    obj.append((1/(2*300))*(np.sum((y_pred-y)**2)))
    theta[0,0]= theta[0,0]-(0.01/300)*(np.sum((y_pred-y)*x))
    theta[0,1]= theta[0,1]-(0.01/300)*(np.sum((y_pred-y)))
```

1000번의 학습후 직선의 모양(red-line)

```python
plt.figure(figsize=(10, 10))
plt.axvline(x=0, color = 'r',linestyle=':')
plt.axhline(y=0, color = 'r',linestyle=':')
plt.title("assignment2-2")
plt.xlabel('X')
plt.ylabel('Y')
t1 = np.arange(-5.0, 5.0, 0.01)
plt.scatter(x,y,color='k',s=3)
plt.plot(t1,theta[0,0]*t1+theta[0,1],color='r')
plt.show()
print("기울기 :",theta[0,0],"y절편:",theta[0,1])
print("원래함수와 거의 같다는것을 알 수 있다.")
```

2-3. Plotting the energy values

```python
plt.figure(figsize=(10, 10))
plt.axvline(x=0, color = 'r',linestyle=':')
plt.axhline(y=0, color = 'r',linestyle=':')
plt.title("assignment2-3(objective function)")
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(range(1000),obj,color='b')
plt.show()
```
