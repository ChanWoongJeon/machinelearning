1. Plot the training data

```python
import numpy as np
import matplotlib.pyplot as plt

data    = np.genfromtxt("data.txt", delimiter=',')

x       = data[:, 0]
y       = data[:, 1]
label   = data[:, 2]

x_label0    = x[label == 0]
x_label1    = x[label == 1]

y_label0    = y[label == 0]
y_label1    = y[label == 1]

plt.figure(figsize=(8, 8))
plt.scatter(x_label0, y_label0, alpha=0.3, c='b')
plt.scatter(x_label1, y_label1, alpha=0.3, c='r')
plt.show()
```

2. Plot the estimated parameters

```python
theta = np.random.randn(1,3)
theta0 = theta[0,0]
theta1 = theta[0,1]
theta2 = theta[0,2]
print(theta0,theta1,theta2)
```

```python
theta_0 = []
theta_1 = []
theta_2 = []
cost_ = []
```

```python
learning_rate = 0.0001
for i in range(10000):
    z = theta0+theta1*x+theta2*y
    sigma_z = 1/(1+np.exp(-z))
    cost = np.sum(-label*np.log(sigma_z)-(1-label)*np.log(1-sigma_z))/len(label)
    theta_0.append(theta0)
    theta_1.append(theta1)
    theta_2.append(theta2)
    cost_.append(cost)
    theta0 = theta0-learning_rate*np.sum(sigma_z-label)/len(label)
    theta1 = theta1-learning_rate*np.sum((sigma_z-label)*x)/len(label)
    theta2 = theta2-learning_rate*np.sum((sigma_z-label)*y)/len(label)
```

```python
plt.figure(figsize=(7, 7))
plt.axvline(x=0, color = 'r',linestyle=':')
plt.axhline(y=0, color = 'r',linestyle=':')
plt.title("assignment5-2(theta)")
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(range(10000),theta_0,color='r')
plt.plot(range(10000),theta_1,color='g')
plt.plot(range(10000),theta_2,color='b')
plt.show()
```
